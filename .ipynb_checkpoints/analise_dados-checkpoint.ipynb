```python
import pandas as pd
import numpy as np
import seaborn as srn
import matplotlib.pyplot as plt
import statistics as sts
```

```python
base = pd.read_csv('dados.csv')
base.head()
```

```python
lista_columns = base['CODIGO;MUNICIPIO;PIB;VALOREMPENHO'].tolist()
```

```python
lista_aux = []
for indice in lista_columns:
    lista_aux.append(indice.split(';'))

lista_columns = lista_aux
```

```python
dados_pd = pd.DataFrame(lista_columns, columns = ['Código', 'Município', 'PIB', 'Empenho'])
dados_pd
```

```python
dados_pd[dados_pd.duplicated(['Código'],keep=False)]
```

```python
dados_pd.drop_duplicates(subset="Código", keep='first',inplace=True)
#buscamos duplicados 
dados_pd[dados_pd.duplicated(['Código'],keep=False)]
```

```python
dados_pd
```

```python
dados_pd.isnull().sum()
```

```python
#Numero de estados com PIB superior a média nacional
#Calculando a média da soma total do PIB pelo numero de municipios
tot_PIB = dados_pd['PIB']
tot_PIB = pd.to_numeric(tot_PIB)
tot_tam = tot_PIB.count()
tot_PIB = tot_PIB.sum()
media = tot_PIB / tot_tam
media
```

```python
PIB = dados_pd['PIB']
PIB = pd.to_numeric(PIB)
menor_med = PIB.loc[(PIB < media)]
maior_med = PIB.loc[(PIB > media)]
igual_med = PIB.loc[(PIB == media)]
maior = maior_med.count()
menor = igual_med.count()
igual = menor_med.count()
```

```python
dados = [maior, menor, igual]
lb_list = ['Acima', 'Igual', 'Abaixo']
dados
```

```python
plt.axis('equal')
plt.pie(dados, labels = lb_list, autopct = '%1.1f%%')
plt.title('PIB Municipal em relação à média do estado')
```
